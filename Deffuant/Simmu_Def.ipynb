## Etude dans le cas d'un graphe complet

```python
N=500
t=2000
G=nx.complete_graph(N) #le graphe
mu=0.5 #capacité d'adaptation
d=0.2
```

```python
M=de.Simulator(G,d,mu,t,conv=True)
```

```python
B=np.arange(len(M))
plt.plot(B,M,"b,")
plt.show()
```

```python
l=len(de.Detecte_pic(M[-1],ndigits=1))
while l<4:
    M=de.Simulator(G,d,mu,t,conv=True)
    l=len(de.Detecte_pic(M[-1]))
B=np.arange(len(M))
plt.plot(B,M,"b,")
plt.show()
```

```python
l=18 #taille du graph
t=30000
G=nx.grid_graph([l,l]) #le graphe
M=nx.adjacency_matrix(G)
G=nx.to_networkx_graph(M)
mu=0.5 #capacité d'adaptation
d=0.3
```

```python
M=de.Simulator_grille(l,d,mu,t)
```

```python
rM=np.reshape(M,newshape=(t,l*l))
B=np.arange(t)
fig=plt.figure(figsize=(20, 10))
plt.plot(B,rM,"b,")
plt.show()
```

# Pic 1 
$l=20$ $\mu$=0.5

```python
l=20 #taille du graph
t=30000
G=nx.grid_graph([l,l]) #le graphe
M=nx.adjacency_matrix(G)
G=nx.to_networkx_graph(M)
mu=0.5 #capacité d'adaptation
```

```python
D=np.linspace(0.05,0.5,20)
M=de.Etude_pics_vect_grille(l,D,mu,t)
Moy=de.Moyenne(M)
V=de.Variance(M)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(D,Moy,marker="x",ls="--", label= "Moyenne")
plt.plot(D,V,marker="x",ls="--", label= "Variance")
plt.title('Nombre moyen de pics en fonction de d')
plt.xlabel("d")
plt.ylabel("nombre moyen de pics")
plt.legend()
plt.grid()
plt.show()
```

```python
pk.dump(M,open("/Users/valentinkilian/Documents/Cours/Stage2021/StageValentin/code/Deffuant/pic1.p","wb"))
```
